```python
import scipy.stats as stats

def chi_squared_test(data, column):
    contingency_table = pd.crosstab(data[column], data['SD_Flag'])
    chi2, p, dof, expected = stats.chi2_contingency(contingency_table)
    print(f"\nChi-Squared Test for '{column}':")
    print("Chi-squared Statistic:", chi2)
    print("Degrees of Freedom:", dof)
    print("P-value:", p)
    if p < 0.05:
        print("Result: Significant association (reject null hypothesis)")
    else:
        print("Result: No significant association (fail to reject null hypothesis)")
    return p
```

```python
def t_test(data, column):
    group1 = data[data['SD_Flag'] == 1][column].dropna()
    group0 = data[data['SD_Flag'] == 0][column].dropna()
    # Check for normality
    stat1, p1 = stats.shapiro(group1)
    stat0, p0 = stats.shapiro(group0)
    print(f"\nShapiro-Wilk Test for Normality on '{column}':")
    print(f"SD Group: Statistic={stat1}, P-value={p1}")
    print(f"NOTSD Group: Statistic={stat0}, P-value={p0}")
    if p1 > 0.05 and p0 > 0.05:
        # Both groups are normally distributed, use t-test
        t_stat, p = stats.ttest_ind(group1, group0, equal_var=False)
        test_type = 'Independent Samples T-Test'
    else:
        # Use Mann-Whitney U Test
        t_stat, p = stats.mannwhitneyu(group1, group0, alternative='two-sided')
        test_type = 'Mann-Whitney U Test'
    print(f"\n{test_type} for '{column}':")
    print("Statistic:", t_stat)
    print("P-value:", p)
    if p < 0.05:
        print("Result: Significant difference (reject null hypothesis)")
    else:
        print("Result: No significant difference (fail to reject null hypothesis)")
    return p
```
