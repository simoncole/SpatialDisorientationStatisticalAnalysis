```python
# Ensure we're working with unique cases based on 'ev_id'
unique_data = data.drop_duplicates(subset='ev_id')

# Total Number of Years Covered
num_years = unique_data['ev_year'].nunique()
print(f"Total Number of Years Covered: {num_years}")

# Total Unique Events Registered
total_unique_events = unique_data.shape[0]
print(f"Total Unique Events Registered: {total_unique_events}")

# Years under investigation
years_investigation = unique_data['ev_year'].sort_values().unique()
print(f"Years under investigation: {years_investigation[0]} - {years_investigation[-1]}")

# Total events in time frame
total_events = unique_data.shape[0]
print(f"Total events in time frame: {total_events}")

# Number of fatal cases
num_fatal_cases = unique_data[unique_data['ev_highest_injury'] == "FATL"].shape[0]
print(f"Number of fatal cases: {num_fatal_cases}")
num_fatal_sd_cases = unique_data[unique_data['Sort'] == 'SD'].shape[0]
print(f"Number of SD-Related cases: {num_fatal_sd_cases}")
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Ensure matplotlib plots display within the Jupyter Notebook
%matplotlib inline

# Increase default figure size for better readability
plt.rcParams['figure.figsize'] = (12, 6)

# Ensure we're working with unique cases based on 'ev_id'
unique_data = data.drop_duplicates(subset='ev_id')

# Split the dataset into SD and NOTSD DataFrames
sd_data = unique_data[unique_data['Sort'] == 'SD']
not_sd_data = unique_data[unique_data['Sort'] != 'SD']

# Function to count unique events for a given column
def count_unique_events(df, column_name, group_name):
    counts = df.groupby(column_name)['ev_id'].nunique().reset_index()
    counts.rename(columns={'ev_id': f'{group_name}_Event_Count'}, inplace=True)
    return counts

# List of columns to analyze
columns_to_analyze = ['Category', 'Subcategory', 'Sections', 'Subsection', 'Who']

# Total unique events in each group
total_sd_events = sd_data['ev_id'].nunique()
total_not_sd_events = not_sd_data['ev_id'].nunique()

# Loop through each column and perform analysis
for col in columns_to_analyze:
    # Count unique events for SD and Non-SD
    sd_counts = count_unique_events(sd_data, col, 'SD')
    not_sd_counts = count_unique_events(not_sd_data, col, 'Non_SD')
    
    # Merge counts for comparison
    comparison = pd.merge(sd_counts, not_sd_counts, on=col, how='outer').fillna(0)
    
    # Calculate percentages
    comparison['SD_Percent'] = (comparison['SD_Event_Count'] / total_sd_events) * 100
    comparison['Non_SD_Percent'] = (comparison['Non_SD_Event_Count'] / total_not_sd_events) * 100
    
    # Sort the comparison DataFrame
    comparison.sort_values(by='SD_Event_Count', ascending=False, inplace=True)
    
    # Limit to top 10 categories for readability (except for 'Category')
    if col != 'Category':
        comparison = comparison.head(10)
    
    # Reset index for plotting
    comparison.reset_index(drop=True, inplace=True)
    
    # Display the comparison DataFrame
    print(f"\nComparison of Findings Profiles at the {col} Level:")
    display(comparison)
    
    # Visualization
    # Set the index to the column for plotting
    comparison.set_index(col, inplace=True)
    
    # Plot the percentages
    ax = comparison[['SD_Percent', 'Non_SD_Percent']].plot(
        kind='barh',
        figsize=(12, 6),
        color=['blue', 'gold']
    )
    ax.set_title(f'Comparison of Event Percentages by {col}')
    ax.set_xlabel('Percentage of Events')
    ax.set_ylabel(col)
    ax.legend(['SD Cases', 'Non-SD Cases'], loc='lower right')
    plt.tight_layout()
    plt.show()
    
    # Reset index for the next iteration
    comparison.reset_index(inplace=True)
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Ensure matplotlib plots display within the Jupyter Notebook
%matplotlib inline

# Increase default figure size for better readability
plt.rcParams['figure.figsize'] = (12, 6)

# Ensure we're working with unique cases based on 'ev_id'
unique_data = data.drop_duplicates(subset='ev_id')

# Split the dataset into SD and NOTSD DataFrames
sd_data = unique_data[unique_data['Sort'] == 'SD']
not_sd_data = unique_data[unique_data['Sort'] != 'SD']

# Function to count unique events for a given column
def count_unique_events(df, column_name, group_name):
    counts = df.groupby(column_name)['ev_id'].nunique().reset_index()
    counts.rename(columns={'ev_id': f'{group_name}_Event_Count'}, inplace=True)
    return counts

# List of columns to analyze
columns_to_analyze = ['Category', 'Subcategory', 'Sections', 'Subsection', 'Who']

# Total unique events in each group
total_sd_events = sd_data['ev_id'].nunique()
total_not_sd_events = not_sd_data['ev_id'].nunique()

# Loop through each column and perform analysis
for col in columns_to_analyze:
    # Count unique events for SD and Non-SD
    sd_counts = count_unique_events(sd_data, col, 'SD')
    not_sd_counts = count_unique_events(not_sd_data, col, 'Non_SD')
    
    # Merge counts for comparison
    comparison = pd.merge(sd_counts, not_sd_counts, on=col, how='outer').fillna(0)
    
    # Calculate percentages
    comparison['SD_Percent'] = (comparison['SD_Event_Count'] / total_sd_events) * 100
    comparison['Non_SD_Percent'] = (comparison['Non_SD_Event_Count'] / total_not_sd_events) * 100
    
    # Calculate the difference between SD and Non-SD percentages
    comparison['Percent_Difference'] = comparison['SD_Percent'] - comparison['Non_SD_Percent']
    
    # Filter where SD_Percent > Non_SD_Percent
    comparison = comparison[comparison['Percent_Difference'] > 0]
    
    # Sort the comparison DataFrame by Percent_Difference in descending order
    comparison.sort_values(by='Percent_Difference', ascending=False, inplace=True)
    
    # Limit to top 10 categories for readability (except for 'Category')
    if col != 'Category':
        comparison = comparison.head(10)
    
    # Reset index for plotting
    comparison.reset_index(drop=True, inplace=True)
    
    # Display the comparison DataFrame
    print(f"\nTop categories where SD percentage is higher in '{col}':")
    display(comparison[[col, 'SD_Percent', 'Non_SD_Percent', 'Percent_Difference']])
    
    # Visualization
    # Set the index to the column for plotting
    comparison.set_index(col, inplace=True)
    
    # Plot the percent differences
    ax = comparison[['SD_Percent', 'Non_SD_Percent']].plot(
        kind='barh',
        figsize=(12, 6),
        color=['blue', 'gold']
    )
    ax.set_title(f'Comparison of Event Percentages by {col} (SD > Non-SD)')
    ax.set_xlabel('Percentage of Events')
    ax.set_ylabel(col)
    ax.legend(['SD Cases', 'Non-SD Cases'], loc='lower right')
    plt.tight_layout()
    plt.show()
    
    # Reset index for the next iteration
    comparison.reset_index(inplace=True)
```
